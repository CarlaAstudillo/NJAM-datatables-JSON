```python
import pandas as pd
```

```python
prevyear = "FY19 \nK to 12 Aid"
latestyear = "FY20 \nK to 12 Aid"
renameprevyear = "aid_18_19"
renamelatestyear = "aid_19_20"
renameaidchange = "aid_change"
renameaidchangeperc = "aid_change_perc"
filename = "district.xlsx"
newfilename = "19_20_aid.csv"
```

```python
school_aid = pd.read_excel(filename,skiprows=4)
```

```python
school_aid.head(5)
```

```python
list(school_aid.columns)
```

```python
school_aid_simple = school_aid[['County','Dist', 'District', prevyear, latestyear, 'FY19- FY20\nK to 12 Aid\nDifference', 'Aid Percent Difference']]
```

```python
school_aid_simple
```

```python
school_aid_simple = school_aid_simple[:-2]
```

```python
school_aid_simple.tail(5)
```

```python
school_aid_simple["Aid Percent Difference Percent"] = school_aid_simple['Aid Percent Difference'] *100
```

```python
school_aid_simple.head(5)
```

```python
list(school_aid_simple.columns)
```

```python
school_aid_simple = school_aid_simple.rename(columns={prevyear: renameprevyear, latestyear: renamelatestyear, 'FY19- FY20\nK to 12 Aid\nDifference': renameaidchange, 'Aid Percent Difference Percent': renameaidchangeperc})
```

```python
school_aid_simple.head(5)
```

```python
school_aid_simple.to_csv(newfilename, index=False)
```

```python
biggest_money_winners = school_aid_simple.sort_values('aid_change', ascending=False)
biggest_perc_winners = school_aid_simple.sort_values('aid_change_perc', ascending=False)
```
